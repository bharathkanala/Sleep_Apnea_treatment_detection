```python
import pandas as pd
import numpy as np

num_cols = [
    'BMI', 'AHI', 'sAHI',
    'mean_o2 [%]', 'O2_sats_below_90 [%]', 'lowest_O2(SS) [%]',
    'RSI', 'NOSE',
    'velopharyngeal_collapse [%]',
    'oropharyngeal_collapse [%]',
    'base_of_tongue_collapse [%]',
    'AirwaySize_mouth_closed', 'AirwaySize_chin_lift'
]

for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print(df[num_cols].dtypes)
```

```python
for col in num_cols:
    if col in df.columns:
        median_value = df[col].median(skipna=True)
        df[col] = df[col].fillna(median_value)
```

```python
df.info()
```

```python
print(df.isna().sum())
```

```python
# Categorical columns (mode)
cat_cols = [
    'tonsils', 'uvula', 'rhinitis', 'DNS',
    'Snoring_mouth_closed', 'Snoring_chin_lift',
    'P', 'T', 'L', 'Tb', 'E'
]

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
print(df.isna().sum())
```

```python
cols = ['STOP_BANG', 'ESS', 'collar_size [cm]']

for col in cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        print(f"✅ {col} converted to numeric. Remaining NaNs: {df[col].isna().sum()}")
    else:
        print(f"Column '{col}' not found in dataframe")

print(df[cols].dtypes)
```

```python
for col in cols:
    if col in df.columns:
        # Calculate median (ignoring NaNs)
        median_value = df[col].median()
        
        # Fill NaNs with median
        df[col] = df[col].fillna(median_value)
        
        print(f"✅ {col} → Filled with median = {median_value:.2f} | Remaining NaNs: {df[col].isna().sum()}")
```

```python
kaggle_df = kaggle_df[[
    'BMI', 'Snoring', 'SpO2',
    'Nasal_Airflow', 'Chest_Movement',
    'AHI'
]]
```

```python
kaggle_df = kaggle_df.copy()
kaggle_df['Snoring'] = kaggle_df['Snoring'].astype(int)
print(kaggle_df['Snoring'].dtype)
```

```python
kaggle_df = kaggle_df.dropna()
```

```python
kaggle_df
```

```python
from sklearn.ensemble import RandomForestRegressor

X_apnoea = kaggle_df[['BMI', 'Snoring', 'SpO2', 'Nasal_Airflow', 'Chest_Movement']]
y_apnoea = kaggle_df['AHI'] * 0.6

model_apnoea = RandomForestRegressor(n_estimators=100, random_state=42)
model_apnoea.fit(X_apnoea, y_apnoea)
```

```python
X_hypopnoea = X_apnoea
y_hypopnoea = kaggle_df['AHI'] * 0.4

model_hypopnoea = RandomForestRegressor(n_estimators=100, random_state=42)
model_hypopnoea.fit(X_hypopnoea, y_hypopnoea)
```

```python
X_flow = kaggle_df[['BMI', 'Snoring', 'SpO2', 'Chest_Movement']]
y_flow = kaggle_df['Nasal_Airflow']

model_flow = RandomForestRegressor(n_estimators=100, random_state=42)
model_flow.fit(X_flow, y_flow)
```

```python
# Create synthetic snore_time
kaggle_df['snore_time'] = kaggle_df['Snoring'] * (1 - kaggle_df['Nasal_Airflow'])

X_snore = kaggle_df[['BMI', 'Snoring', 'SpO2', 'Nasal_Airflow']]
y_snore = kaggle_df['snore_time']

model_snore = RandomForestRegressor(n_estimators=100, random_state=42)
model_snore.fit(X_snore, y_snore)
```
