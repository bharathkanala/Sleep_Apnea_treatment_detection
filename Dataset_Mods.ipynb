```python
selected_columns = [

    #Core physiological
    'BMI', 'AHI', 'sAHI',
    'mean_o2 [%]', 'O2_sats_below_90 [%]', 'lowest_O2(SS) [%]',
    'apnoea', 'hypopnoea',

    #Airway/breathing
    'RSI', 'NOSE',
    'Snoring_mouth_closed', 'Snoring_chin_lift',
    'AirwaySize_mouth_closed', 'AirwaySize_chin_lift',

    #Anatomy (VERY IMPORTANT for treatment)
    'tonsils', 'uvula', 'rhinitis', 'DNS',
    'velopharyngeal_collapse [%]',
    'oropharyngeal_collapse [%]',
    'base_of_tongue_collapse [%]',

    #Optional
    'collar_size [cm]',
    'STOP_BANG',
    'ESS',
    'snore_time',
    'Flow',
    'dB',

    #DISE scoring (very useful for treatment)
    'P', 'T', 'L', 'Tb', 'E',

    #TARGET
    'treatment_plan_with_stages'
]
```

```python
df = df[selected_columns]
```

```python
df.to_csv("ReqData.csv", encoding='utf-8', index=False)
```

```python
df = df.replace('', pd.NA)
df = df.dropna(how='all').reset_index(drop=True)
```

```python
df = df[df['treatment_plan_with_stages'].notna()]
```

```python
missing = df.isnull().sum().sort_values(ascending=False)
print(missing)
```

```python
categorical_cols = ['tonsils', 'uvula', 'rhinitis', 'DNS', 
                    'Snoring_mouth_closed', 'Snoring_chin_lift']

for col in categorical_cols:
    if col in df.columns:
        # Convert NaN and floats to proper string representation first
        df[col] = df[col].replace([np.nan, None], pd.NA)
        df[col] = df[col].astype(str).str.strip().replace('nan', pd.NA)
        
        # Now safely convert to category
        df[col] = df[col].astype('category')
```
